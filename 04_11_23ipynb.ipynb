```python
# series and dataframe
# descriptive analytics
import matplotlib.pyplot as plt
# nominal ,ordinal ,interval and ratio
```

```python
from matplotlib.lines import MarkerStyle
import numpy as np
x = np.array([25,50,100,125])
y = np.array([1,2,3,4])
plt.plot(x,y,marker = 'o',markersize= 7,markerfacecolor = "#ffffff")
x1 = np.array([25,55,70,90])
y1 = np.array([2,4,6,7])
plt.plot(x1,y1,marker = 'o',markersize = 7,markerfacecolor = "#676877")
plt.show()
# linestyle
```

```python
plt.savefig("xytest.png",dpi=300)
```

```python
plt.plot(x,y,"o",ls='-')
plt.plot(x,y1,"o",ls="--")
plt.xlabel("x label")
plt.ylabel("y1 label")
plt.title("title of the chart")
plt.grid(c = "#345678")
plt.show()
```
