```python
# ploynomial regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array(range(1,41))
```

```python
y = x*x*x*x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
x = x.reshape((-1,1))
y = y.reshape((-1,1))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 2022)
```

```python
print("shape of training input",x_train.shape)
print("shape of training output",y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lg_regression = LinearRegression()
lg_regression.fit(x_train,y_train)
```

```python
y_pred = lg_regression.predict(x_test)
```

```python
plt.plot(x,lg_regression.predict(x))
plt.plot(x,y)
```

```python
# what is the use and advantage of polynomial regression ?////
y [5] = 23
y [-2] = 13441
plt.plot(x,lg_regression.predict(x))
plt.plot(x,y)
```

```python
# increasing in degrees too much - overfit solution
# decreasing in degrees too much - underfit solution
# when we are using polynomial regression it takes all the features as degrees
# if the feature value is 3 then degree as 3
# intercept is called - bias
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_feat = PolynomialFeatures(degree = 2) # this is not ml model
# we use only fit and transform the object
# transform the model(object)
```

```python
x_degree_two = poly_feat.fit_transform(x)
```

```python
print(x)
```

```python
print(x_degree_two)
```

```python
poly_model = LinearRegression()
poly_model.fit(x_degree_two,y)
plt.plot(x,y)
plt.plot(x,poly_model.predict(x_degree_two))
plt.plot(x,lg_regression.predict(x))
```

```python
# linear regression is under fit model
# multiple linear regression is over fit model
poly_feat = PolynomialFeatures(degree = 40)
x_degree = poly_feat.fit_transform(x)
poly_model = LinearRegression()
poly_model.fit(x_degree,y)
plt.plot(x,y)
plt.plot(x,poly_model.predict(x_degree))
poly_model.score(x_degree,y)
#plt.plot(x,lg_regression.predict(x))
```

```python
import pandas as pd
land_ocean_df = pd.read_csv("/content/global-land-oceantempindex-copy.csv",
                            sep ='    ',
                            skiprows=6,
                            names = ['year','No_smoothing','Smoothing_temp'])
```
